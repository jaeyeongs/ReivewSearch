```python
import pandas as pd
```

## 리뷰 데이터 불러오기

```python
# 웅이네오돌뼈닭발도
df = pd.read_csv('data.csv', sep = ',')
df
```

```python
df.isnull().sum() # NULL 값 유무 확인
```

```python
df = df.dropna(axis=0) # NULL 값이 있는 행 제거
print(df.isnull().sum()) # NULL 값 유무 재확인
```

```python
# 분석에 필요한 변수만 불러오기
df = df.loc[:, ["category", "store", "id", "review_spell_check", "star"]]
df
```

## 리뷰 검색 모듈

```python
user_input = input('검색 : ')

search_df = df[df['review_spell_check'].str.contains(user_input)]

if len(search_df) == 0:
    print("검색 결과가 없습니다.")
else:
    for idx, category, store, user_id, review, star in search_df.itertuples():
        print("[{}] [{}] {}의 평점: {}점\n리뷰: {}".format(
            category, store, user_id, star, review))
        
```
